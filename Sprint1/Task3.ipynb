В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма.

#### Дополнительное задание
Добавьте к сущестующим классам конструктора классов с использованием гетторов и сетторов и реализуйте взаимодействие объектов между собой

```
using System;

class Customer
{
    private int customerId;
    private string name;
    private string email;

    public int CustomerId
    {
        get { return customerId; }
        set { customerId = value; }
    }

    public string Name
    {
        get { return name; }
        set { name = value; }
    }

    public string Email
    {
        get { return email; }
        set { email = value; }
    }

    public Customer(int id, string name, string email)
    {
        this.CustomerId = id;
        this.Name = name;
        this.Email = email;
    }

    public virtual string GetFullName()
    {
        return Name;
    }

    public virtual void UpdateEmail(string newEmail)
    {
        Email = newEmail;
        Console.WriteLine($"Email обновлён на: {newEmail}");
    }

    public virtual void ViewProfile()
    {
        Console.WriteLine($"ID клиента: {CustomerId}, Имя: {GetFullName()}, Электронная почта: {Email}");
    }
}

class VipCustomer : Customer
{
    private int loyaltyPoints;

    public int LoyaltyPoints
    {
        get { return loyaltyPoints; }
        set { loyaltyPoints = value; }
    }
    
    public VipCustomer(int id, string name, string email, int loyaltyPoints) : base(id, name, email)
    {
        this.LoyaltyPoints = loyaltyPoints;
    }

    public override void ViewProfile()
    {
        base.ViewProfile();
        Console.WriteLine($"Баланс лояльности: {LoyaltyPoints}");
    }
}

class RegularCustomer : Customer
{
    private string registrationDate;

    public string RegistrationDate
    {
        get { return registrationDate; }
        set { registrationDate = value; }
    }
    
    public RegularCustomer(int id, string name, string email, string registrationDate) : base(id, name, email)
    {
        this.RegistrationDate = registrationDate;
    }

    public override void UpdateEmail(string newEmail)
    {
        base.UpdateEmail(newEmail);
        Console.WriteLine($"Дата последнего обновления электронной почты: {DateTime.Now}");
    }
}

class GroupCustomer : Customer
{
    private string groupName;
    private List<RegularCustomer> customers = new List<RegularCustomer>();

    public string GroupName
    {
        get { return groupName; }
        set { groupName = value; }
    }

    public GroupCustomer(int id, string groupName, string email) : base(id, groupName, email)
    {
        this.GroupName = groupName;
    }

    public void Add(RegularCustomer customer)
    {
        customers.Add(customer);
    }


    public override string GetFullName()
    {
        return GroupName; 
    }

    public override void ViewProfile()
    {
        Console.WriteLine($"ID группы клиента: {CustomerId}, Название группы: {GetFullName()}, Электронная почта: {Email}");
        Console.WriteLine("Клиенты в группе:");
        foreach (var customer in customers)
        {
            customer.ViewProfile();
        }
    }
}

VipCustomer vipCustomer = new VipCustomer(1, "Александр Пупкин", "Aleksandr_Pupkin@mail.ru", 150);
Console.WriteLine("Профиль VIP клиента:");
vipCustomer.ViewProfile();

RegularCustomer regularCustomer1 = new RegularCustomer(2, "Юлия Зайцева", "Julia_Zaitseva@mail.ru", DateTime.Now.ToString("dd/MM/yyyy"));
Console.WriteLine("\nПрофиль обычного клиента:");
regularCustomer1.ViewProfile();
regularCustomer1.UpdateEmail("Julia_Zaitseva@mail.ru");

RegularCustomer regularCustomer2 = new RegularCustomer(3, "Глеб Всезнайка", "vceznaika@mail.ru", DateTime.Now.ToString("dd/MM/yyyy"));
Console.WriteLine("\nПрофиль обычного клиента:");
regularCustomer2.ViewProfile();
regularCustomer2.UpdateEmail("vceznaika@mail.ru");

GroupCustomer groupCustomer = new GroupCustomer(4, "Группа A", "group_a@mail.ru");

groupCustomer.Add(regularCustomer1);
groupCustomer.Add(regularCustomer2);

Console.WriteLine("\nПрофиль группового клиента:");
groupCustomer.ViewProfile();

static void InteractCustomers(Customer customer1, Customer customer2)
{
    
    Console.WriteLine($"Клиент {customer1.GetFullName()} отправляет привет клиенту {customer2.GetFullName()}.");
    Console.WriteLine($"Клиент {customer2.GetFullName()} отвечает: Здравствуй, {customer1.GetFullName()}!");
}
Console.WriteLine("\nВзаимодействие между клиентами:");
InteractCustomers(vipCustomer, regularCustomer1);
```
